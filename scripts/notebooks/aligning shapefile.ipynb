```python
import os
import ogr
import osr
import csv
import json
import numpy as np
driverName = "ESRI Shapefile"
driver = ogr.GetDriverByName(driverName)
```

```python
poly_file = './rotated-test.shp'
point_file = './test-utm-madhav.shp'
data_poly = driver.Open(poly_file)
data_point = driver.Open(point_file)
```

```python
poly_layer = data_poly.GetLayerByIndex(0)
point_layer = data_point.GetLayerByIndex(0)
```

```python
output_poly = './test_out.shp'
if os.path.exists(output_poly):
    print("Specified output file already exists.\n"\
    "Deleting current file and replaing it with new points file")
    driver.DeleteDataSource(output_poly)
output_data = driver.CreateDataSource(output_poly)
out_srs = osr.SpatialReference()
out_srs.ImportFromEPSG(32643)
out_layer = output_data.CreateLayer("poly_layer",out_srs,ogr.wkbLineString)
field_id = ogr.FieldDefn("id",ogr.OFTInteger)
out_layer.CreateField(field_id)
```

```python
count = 0
for i in range(poly_layer.GetFeatureCount()):
    ftr_in = poly_layer[i]
    geom = ftr_in.GetGeometryRef()
    geomJSON = json.loads(geom.ExportToJson())
    payload = {'type':'LineString'}
    points = np.array(geomJSON['coordinates'])
    points_new = points/1000.0 + c
    payload['coordinates'] = points_new.tolist()
    
    ftr_out = ogr.Feature(out_layer.GetLayerDefn())
    ftr_out.SetField("id",count)
    count += 1
    poly = ogr.CreateGeometryFromJson(json.dumps(payload))
    print(poly.ExportToJson())
    ftr_out.SetGeometry(poly)
    out_layer.CreateFeature(ftr_out)
    ftr_out = None
    
    
    
    
    
    
```
